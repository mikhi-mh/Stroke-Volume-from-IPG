#Test , Test Split


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=43)
```

#left intension

#Linear Regression

```python
from sklearn import linear_model
model_linear = linear_model.LinearRegression()
```

```python
model_linear.fit(x_train, y_train)
```

```python
y_pred_linear = model_linear.predict(x_test)
```

```python
y_pred_linear
```

```python
print(y_pred_linear.shape)
```

```python
#score Model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
print("Mean Absolute Error is : {}".format( mean_absolute_error(y_test,y_pred_linear) ) )

print("Root Mean Squared Error is : {} ".format(np.sqrt(mean_squared_error(y_test,y_pred_linear))))
```

#AdaBoost


```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
model_adaBoost = AdaBoostRegressor()
```

```python
model_adaBoost.fit(x_train,y_train)
```

```python
y_pred_adaBoost = model_adaBoost.predict(x_test)
```

```python
y_pred_adaBoost
```

```python
print( "Mean Absolute Error is : {}".format(mean_absolute_error(y_test,y_pred_adaBoost) ) )

print("Root Mean Squared Error is : {} ".format(np.sqrt(mean_squared_error(y_test,y_pred_adaBoost))))
```

#Gradient Boost


```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model_gradient = GradientBoostingRegressor(random_state = 0)
```

```python
model_gradient.fit(x_train,y_train)
```

```python
y_pred_gradient = model_gradient.predict(x_test)
```

```python
y_pred_gradient
```

```python
y_pred_gradient
```

```python
print( "Mean Absolute Error is : {}".format(mean_absolute_error(y_test,y_pred_gradient) ) )

print("Root Mean Squared Error is : {} ".format(np.sqrt(mean_squared_error(y_test,y_pred_gradient))))
```
