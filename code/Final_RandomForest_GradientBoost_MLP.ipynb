#MLP

```python
from sklearn.neural_network import MLPRegressor
```

```python
model_MLP = MLPRegressor(
    hidden_layer_sizes = (30,30,30),
    activation = "logistic",
    random_state = 0
).fit(x_train, y_train)
```

#MLP2

```python
from sklearn.neural_network import MLPRegressor
```

```python
model_MLP = MLPRegressor(random_state=1, max_iter=3000).fit(x_train, y_train)
print('The R2 score of our prediction is %.4f' %(model_MLP.score(x_test, y_test)))

y_pred_MLP = model_MLP.predict(x_test)
```

```python
y_pred_MLP
```

# Hybrid


```python
from mlxtend.regressor import StackingRegressor
from sklearn.svm import SVR
```

```python
regressors = [model_linear, model_gradientBoost, model_MLP , model_randomForest,model_adaBoost,model_linear]
# regressors = [model_lr, model_nnr]
svr_rbf = SVR(kernel = 'rbf')

stregr = StackingRegressor(
    regressors = regressors, 
    meta_regressor = svr_rbf
).fit(x_train, y_train)

pred_stack = stregr.predict(x_test)
```

```python
pred_stack
```

#Error Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
print("Mean Absolute Error is : ")
mean_absolute_error(y_test,pred_stack)
```

```python
print("Mean Squared Error is ")
mean_squared_error(y_test,pred_stack)
```

```python
print("Root Mean Squared Error is ")
np.sqrt(mean_squared_error(y_test,pred_stack))
```
